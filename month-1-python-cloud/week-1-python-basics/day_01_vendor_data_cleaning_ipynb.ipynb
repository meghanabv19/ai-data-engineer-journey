```python
# Vendor Data Cleaning Pipeline
# Translating DataStage NULL handling & type coercion logic to Python
# Pattern: data standardisation, type safety, invalid record filtering
# Meghana BV — github.com/meghanabv19

raw_vendors = [
    {"id": "V001", "name": "  Acme Corp ", "country": "uk", "revenue": "15000", "active": "yes"},
    {"id": "V002", "name": "GlobalTech", "country": "US", "revenue": None, "active": "YES"},
    {"id": "V003", "name": "   ", "country": "india", "revenue": "9500", "active": "no"},
    {"id": "V004", "name": "Synergy Ltd", "country": "UK", "revenue": "20000", "active": "yes"},
    {"id": "V005", "name": "  Delta Systems  ", "country": "us", "revenue": "abc", "active": "YES"}
  ]

# Helper function to summarise vendors, defined globally for clear scope
def summarise_vendors_result(vendors_data):
  total_vendors = len(vendors_data)
  active_vendors = sum(1 for vendor in vendors_data if vendor['active'])
  total_revenue = sum(vendor['revenue'] for vendor in vendors_data)
  print(f"Total vendors: {total_vendors}")
  print(f"Active vendors: {active_vendors}")
  print(f"Total revenue: {total_revenue}")

def clean_vendors(vendors_to_clean):
  cleaned_vendors_list = [] # List to store valid and cleaned vendor data

  for vendor_dict in vendors_to_clean:
    current_vendor = vendor_dict.copy() # Work on a copy of the dictionary

    # Remove spaces from name
    current_vendor['name'] = current_vendor['name'].strip()

    # Uppercases `country`
    current_vendor['country'] = current_vendor['country'].upper()

    # Converts `revenue` to integer — if it's `None` or not a valid number, set it to `0`
    try:
        current_vendor['revenue'] = int(current_vendor['revenue'] or 0)
    except (ValueError, TypeError):
        current_vendor['revenue'] = 0

    # Converts `active` to a proper boolean (`True`/`False`)
    active_value = current_vendor['active']
    current_vendor['active'] = str(active_value).strip().lower() == 'yes'

    # **Removes** any vendor where `name` is empty or only whitespace
    if not current_vendor['name'] :
      # If name is invalid, this vendor is skipped and not added to cleaned_vendors_list
      continue
    else:
      cleaned_vendors_list.append(current_vendor)


  # Summarise the cleaned vendors
  summarise_vendors_result(cleaned_vendors_list)
  return(cleaned_vendors_list)
# Call the cleaning function with the global raw_vendors list
clean_vendors(raw_vendors)
```
